## Population analysis

```python
proc = SpikeTrainAnalysis(df_spikes, df_stimulus, df_correct, stim_list=[0, 45, 90, 135, 180, 225, 270, 315], 
                start_fixation=0, end_fixation=1000, start_delay=1750, end_delay=4750, binsize=250)
allneurons = proc.allneurons  # list of all neurons' names
FF_fixation = []
FF_delay_pref = []
FF_delay_least_pref = []
dFF_pref = []
dFF_least_pref = []
bin_FF_pref = np.array([])  # delay - foreperiod for (neuron,bin) pairs
bin_FF_least_pref = np.array([])  # delay - foreperiod for (neuron,bin) pairs
```

```python
for neuron in allneurons:
    #print(neuron)
    _, _, _, is_tuned = proc.fit(neuron, start=1750, end=4750, binsize=250, ANOVA=True)
    if is_tuned:
        FF_delay_preferred, FF_delay_least_preferred, FF_foreperiod, is_tuned, corFRFF, corFRFF_p, stim_preferred, stim_least_preferred, bin_FF_delay_preferred, bin_FF_delay_least_preferred = proc.single_neuron_stat(neuron)
        FF_fixation.append(FF_foreperiod)
        FF_delay_pref.append(FF_delay_preferred)
        FF_delay_least_pref.append(FF_delay_least_preferred)
        dFF_pref.append(FF_delay_preferred - FF_foreperiod)
        dFF_least_pref.append(FF_delay_least_preferred - FF_foreperiod)
        bin_FF_pref = np.append(bin_FF_pref, bin_FF_delay_preferred - FF_foreperiod)
        bin_FF_least_pref = np.append(bin_FF_least_pref, bin_FF_delay_least_preferred - FF_foreperiod)
```

```python
plt.rc('font', size=24)
plt.rc('axes', autolimit_mode='round_numbers', xmargin=0, ymargin=0)
plt.rc('legend', fontsize='x-small')
plt.rc('xtick', direction='out')
plt.rc('ytick', direction='out')
```

```python
bins_arr = np.arange(0,6,0.25)
plt.figure(figsize=(8,6))
ax = plt.gca()
n, bins, patches = ax.hist(FF_fixation, range=[0, 6],histtype='stepfilled', linewidth = 4, fc = (216.0/255,179.0/255,101.0/255, 0.3), edgecolor = (216.0/255,179.0/255,101.0/255),bins=bins_arr,label='Foreperiod')
n, bins, patches = ax.hist(FF_delay_pref, range=[0, 6],histtype='stepfilled', linewidth = 4, fc = (145.0/255,120.0/255,215.0/255, 0.3), edgecolor= 'mediumpurple', bins=bins_arr,label='Delay preferred') # (145.0/255,191.0/255,219.0/255)
n, bins, patches = ax.hist(FF_delay_least_pref,range=[0, 6], histtype='stepfilled', linewidth = 4, fc = (145.0/255,207.0/255,96.0/255, 0.3), edgecolor = (145.0/255,207.0/255,96.0/255), bins=bins_arr,label='Delay least preferred')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_linewidth(2.0)
ax.spines['bottom'].set_linewidth(2.0)
ax.tick_params(top=False, right=False, length=10.0,width=2.0)
plt.legend(loc='best',frameon=False)
plt.xlabel('Fano factor (FF)',fontsize=24)
plt.ylabel('Number of neurons',fontsize=24)
plt.tight_layout()
plt.show()
```

```python
bins_arr = np.arange(-6,6,0.25)
plt.figure(figsize=(8,6))
ax = plt.gca()
ax.hist(bin_FF_pref[~np.isnan(bin_FF_pref)], histtype='stepfilled', linewidth = 4, fc = (145.0/255,120.0/255,215.0/255, 0.3), edgecolor= 'mediumpurple',bins=bins_arr,label='Preferred') # (145.0/255,191.0/255,219.0/255)
ax.hist(bin_FF_least_pref[~np.isnan(bin_FF_least_pref)], histtype='stepfilled', linewidth = 4, fc = (145.0/255,207.0/255,96.0/255, 0.3), edgecolor = (145.0/255,207.0/255,96.0/255), bins=bins_arr,label='Least preferred')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_linewidth(2.0)
ax.spines['bottom'].set_linewidth(2.0)
ax.tick_params(top=False, right=False, length=10.0,width=2.0)
#ax.xaxis.set_ticks(np.arange(-3.0,2.5,1.0))
plt.axvline(x=0.0, color = 'r',linewidth=4)
plt.legend(loc='best',frameon=False)
plt.xlabel('FF (Delay) - FF (Foreperiod)',fontsize=24)
plt.ylabel('Number of (neuron, bin) pairs',fontsize=20)
plt.tight_layout()
plt.show()
```
